# Smart Attendance & Study Planner

This project helps students calculate attendance percentage, predict required classes to reach 75%, and plan study time effectively.

## Problem Statement

Students find it difficult to track attendance and ensure they meet the 75% requirement.

## Objective

- Calculate attendance percentage  
- Predict required classes  
- Generate study plan  
- Visualize attendance  

## Conclusion

The program successfully calculates attendance, predicts required classes, and provides a simple study plan.

```python
import matplotlib.pyplot as plt

# calculate attendance percentage
def calculate_attendance(attended, total):
    if total == 0:
        return 0
    return (attended / total) * 100


#  calculate classes needed for 75%
def classes_needed_for_75(attended, total):
    required = 0
    while (attended / total) * 100 < 75:
        attended += 1
        total += 1
        required += 1
    return required


#  study planner
def study_planner(subjects, total_hours):
    plan = {}
    per_subject = total_hours / len(subjects)

    for subject in subjects:
        plan[subject] = round(per_subject, 2)

    return plan


#  plot attendance graph
def plot_attendance(attended, total):
    labels = ['Attended', 'Missed']
    values = [attended, total - attended]

    plt.pie(values, labels=labels, autopct='%1.1f%%')
    plt.title("Attendance Distribution")
    plt.show()


# ---------------- MAIN PROGRAM ----------------

print("=== Smart Attendance & Study Planner ===")

total = 30
attended = 18

print(f"Total Classes: {total}")
print(f"Classes Attended: {attended}")

# Attendance calculation
percentage = calculate_attendance(attended, total)
print(f"\nAttendance Percentage: {percentage:.2f}%")

# Classes needed
needed = classes_needed_for_75(attended, total)
print(f"Classes needed to reach 75%: {needed}")

# Graph
plot_attendance(attended, total)

# Study planner
print("\n--- Study Planner ---")

subjects = ["Physics", "Chemistry", "Mathematics"]
hours = 6

plan = study_planner(subjects, hours)

for subject, time in plan.items():
    print(f"{subject}: {time} hours")
```
